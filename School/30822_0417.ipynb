2. 예시:
 * 군집화 알고리즘 (고객 성향 파악, 추천 시스템 등의 군집화를 진행)
3. 분류와 군집화의 차이점:
 * 정답이 있는 상태에서 구별하면 분류 알고리즘, 정답이 없는 상태에서 일단 나눠지면 군집화라고 한다.
