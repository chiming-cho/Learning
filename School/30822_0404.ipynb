```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df=pd.read_csv("/content/cloudiness03_data_01.csv",encoding='cp949')
plt.scatter(df['평균 전운량(1/10)'], df['합계 일조 시간(hr)'],color='black')
plt.xlabel("avg total cloud amount")
plt.ylabel("total sunrise time")
plt.show()
```

```python
print(df["합계 일조 시간(hr)"])
```

```python
#피어슨 상관계수
print("피어슨 상관계수: ", df['평균 전운량(1/10)'].corr(df['합계 일조 시간(hr)'], method="pearson"))

# 변수들이 얼마나 직선형으로 연관되어 있는지를 바탕으로 상관관계를 파악한다.
#corr: correlation function (which performs pearson correlation coefficient, it measures the linear relationship between pairs of variable)
# +1 means perfect Positive correlation.
# 0 means no correlation
# -1 means perfect Negative correlation
```

```python
# 스피어만 상관계수
print("피어슨 상관계수: ", df['평균 전운량(1/10)'].corr(df['합계 일조 시간(hr)'], method="spearman"))

#두 변수 간의 순위를 측정하여 상관관계를 파악한다.
```

```python
from sklearn.linear_model import LinearRegression

x=np.array(df["평균 전운량(1/10)"]).reshape(-1,1)
y=np.array(df["합계 일조 시간(hr)"]).reshape(-1,1)

print(x)
model=LinearRegression().fit(x,y)
print("기울기: ",model.coef_, "절편: ",model.intercept_)
```

```python
slope=model.coef_[0]
intercept=model.intercept_

plt.scatter(x,y,color='green',label='scatter')
plt.plot(x,(slope*x+intercept),color='red',label='regression')

plt.xlabel('avg total cloud amount')
plt.ylabel('total sunrise time')
plt.title('Linear Regression: tot clound amt and tot sunrise time')
plt.legend()
plt.show()
```

```python
df2=pd.read_csv("/content/population_data_02.csv",encoding='cp949')
plt.scatter(df2["총계 남"],df2["총계 여"],color='gold')
plt.xlabel("total man population")
plt.ylabel("total woman population")
plt.show()

print("피어슨 상관계수: ", df2['총계 남'].corr(df2['총계 여'], method="pearson"))
print("피어슨 상관계수: ", df2['총계 남'].corr(df2['총계 여'], method="spearman"))
```

```python
x=np.array(df2["총계 남"]).reshape(-1,1)
y=np.array(df2["총계 여"]).reshape(-1,1)

# print(x)
model=LinearRegression().fit(x,y)

slope=model.coef_[0]
intercept=model.intercept_

plt.scatter(x,y,color='gold',label='scatter')
plt.plot(x,(slope*x+intercept),color='red',label='regression')
plt.xlabel("total man population")
plt.ylabel("total woman population")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
