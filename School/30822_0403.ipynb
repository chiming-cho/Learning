```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/data_02.csv',encoding='cp949')
plt.scatter(range(1,32),df['평균기온(°C)'])
plt.xlabel('date')
plt.ylabel('avg temp(°C)')
plt.title("avg temp in daegu")
```

```python
print("범위: ", max(df['평균기온(°C)'])-min(df['평균기온(°C)']))
print("사분위간 범위: ", df['평균기온(°C)']. quantile(0.75)-df['평균기온(°C)'].quantile(0.25))
```

```python
print("분산: ", df['평균기온(°C)'].var())
print("표준편차: ", df['평균기온(°C)'].std())
```

```python
plt.hist(df['평균기온(°C)'],color='skyblue')
```

```python
from scipy import stats
import seaborn as sns
import numpy as np

sns.distplot(df['평균기온(°C)'],fit=stats.norm)
```
