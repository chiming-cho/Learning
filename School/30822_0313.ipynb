예제 09

```python
# ex 09
s = 0
for i in range(1,101):
    s = s + i
print("1에서 100까지의 합: ",s)

i, s = 1, 0
while i<101:
    s = s + i
    i = i + 1
print("1에서 100까지의 합: ",s)
```

예제 10

```python
# ex 10
def total (a,b,c):
    sum = a+b+c
    return sum
kor = int(input("국어: "))
eng = int(input("영어: "))
mat = int(input("수학: "))
hap = total(kor, eng, mat)
avg = hap/3
print("총합: ",hap, ",", "평균: ",avg)
```
