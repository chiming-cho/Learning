```python
import folium
m = folium.Map(location=[37.5838699,127.0565831], zoom_start=15)
folium.Marker(
location = [37.5, 126.9],
popup = 'Seoul City Hall',
icon = folium.Icon(color='red', icon='info-sign')).add_to(m)
m.add_child(folium.ClickForMarker(popup='Here'))
m.save('/map.html')
```
