##**단일 선형회귀**
* 선형 회귀를 1회 적용하여 데이터를 에측한다.

```python
import numpy as np
x=np.array([[74],[81],[94],[68],[64],[70],[83],[94],[71],[74]]) #모의고사
y=np.array([71,78,90,68,66,72,89,92,78,80])                     #수능
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.scatter(x,y,c='b')
plt.xlabel('exam')
plt.ylabel('final')
plt.show()

print('상관계수', np.corrcoef(x.reshape(len(y)),y))
# x와 y의 상관계수가 표의 형태로 나오게 된다. 각각 가로줄은 x,y, 세로줄도 x,y 이다. 따라서 x와 x, y와 y는 1인 것을 볼 수 있다.
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression(fit_intercept=True)
```

```python
model=lin_reg.fit(x,y)

w=model.coef_ #기울기를 w로 표현함, 그 이유는 가중치 weight 이기 때문
b=model.intercept_ #바이오스 bios (절편)
print("w=%.4f, b=%.4f"%(w,b))
r2=model.score(x,y)
print("r^2=%.4f"%(r2))
```

```python
plt.plot(x,y,'bo',label='real data') # 실젯값으로 산점도 그리기
plt.plot(x,np.array(w*x+b), 'r',label='prediction') # 모델의 예측값으로 선그래프 그리기
plt.xlabel('exam')
plt.ylabel('final')
plt.legend()
plt.show()
```

```python
ex_new=np.array([[80],[90]])
print(model.predict(x_new)) #model=lin_reg.fit(x,y)
```
