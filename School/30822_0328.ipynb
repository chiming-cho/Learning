```python
import numpy as np
a=np.array([1,2,3,4,5])
avg=np.mean(a)
print("평균값: ",avg)
```

```python
import random as rd

li=[]

for i in range(5):
  li.append(rd.randint(1,10))
print("li: {}".format(li))

arr=np.array(li)
print("arr: {}".format(arr))

avg=np.mean(li)
print("avg: ",avg)
```

```python
import pandas as pd
df=pd.read_csv('/data_01.csv',encoding='cp949')
print(df)
print()
print("average: {:.2f}".format(df['평균기온(°C)'].mean()))
print("median: {}".format(df['평균기온(°C)'].quantile(0.5)))
print("mode :", df['평균기온(°C)'].mode()[0])

print(pd.value_counts(df['평균기온(°C)']))
# pd.Series(obj).value_counts()
```
