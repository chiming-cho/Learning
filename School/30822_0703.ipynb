## **데이터 분석과 예측**


## 1. 머신러닝의 정의
- 머신러닝이란 컴퓨터에게 명시적으로 프로그램을 하지 않고 컴퓨터가 학습할 수 있는 능력을 갖는 것을 말한다.
- 컴퓨터가 어떤 작업을 하는 데 있어서 경험으로부터 학습하여 성능을 향상시키는 기술을 머신러닝이라고 정의할 수 있다.

## 2. 머신러닝을 활용하기에 적절하지 않은 문제는?
- 1. 지속적으로 조건이 변화하는 문제인 경우
