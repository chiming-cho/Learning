```python
from numpy.random import randn
np.random.seed(1004)
data=randn(5,4)
data
```

```python
df=pd.DataFrame(data,
                index='A B C D E'.split(),
                columns='가 나 다 라'.split())
df
```

```python
df['나']
```

```python
df[['나','다']]
```

```python
df['합']=df['나']+df['라']
df
```

```python
df.drop('합',axis=1,inplace=True)
df
```

```python
df.drop('E',axis=0,inplace=True)
df
```

```python
df[df>1]
df
```
