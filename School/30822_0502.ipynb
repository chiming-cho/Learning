##**이진 분류 모델**
* 입력을 두 개의 클래스로 분류하는 모델이다.
* 입력에 따라 나타나는 출력이 연속적인 값이 아닐 경우 분류 모델을 사용한다.

#### 로지스틱 회귀

* 선형 회귀와는 달리 두 가지 클래스로 나누는데 적합하다.
* 선형회귀는 일차함수로 나오기 때문에 적합하지 않지만, 로지스틱 회귀는 시그모이드 곡선을 사용하여 두 가지 클래스로 분류하는데 적합하다.
* 시그모이드 함수에서는 출력값 H(x)가 0.5 이상 또는 미만으로 분류하여 1과 0으로 나눈다.
* 분류는 정확한 것이 아니라 확률적으로 나누는 것이다. ~~양자컴퓨터?~~

#### **이진 분류**

```python
import numpy as np
x = np.array([[74],[81],[94],[68],[64],[70],[83],[94],[71],[74]]) #시험 점수
y = np.array([1, 1, 1, 0, 0, 0, 1, 1, 0, 1]) #합불여부

from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='lbfgs')

model=log_reg.fit(x,y)

w=model.coef_
b=model.intercept_
print("w=",w,"b=",b) #로지스틱 회귀에서는 자연상수 e의 지수에 올라가는 wx+b 형태의 식을 먼저 구한다.
```

```python
def H(x):
  return 1/(1+np.exp(-x))
```

```python
x_new=np.array([[72],[73]]) #새로운 시험점수
print(H(x_new*w+b))
print("예측 결과 = ", model.predict(x_new)) # 예측결과
```

```python
# 그래프로 시각화하기
x_in = np.arange(x.min(),x.max(),0.5)
x_in=x_in.reshape(-1,1) # x를 2차원 배열로 형 변환
Hx = H(w*x_in+b) # 선형 회귀의 출력을 H(x)의 입력으로 하기
import matplotlib.pyplot as plt # 그래프 그리기
plt.plot(x, y, 'bo', label='real data') # 실제 데이터
# H(x) sigmoid 함수의 출력
plt.plot(x_in, Hx, 'g', label='H(x)')
# 모델의 예측
plt.plot(x, model.predict(x),'rx', label='prediction')
plt.xlabel("exam score")
plt.ylabel("result")
plt.legend(loc='right')
plt.grid(True)
plt.show()
```
