## **각종 입력**
* 날짜와 배운 내용 입력
* 현재 파일 상태 확인

```python
# 배운 날짜와 배운 내용 입력받아서 daylist.csv 파일에 추가하기
from IPython.display import display
tmp=pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/daylist.csv')

while True: # while 문으로 반복실행 없이 계속 추가 가능하도록 구현
    day=input("Input Date (format: MMDD): ") # 학습 날짜 입력

    if day=='stop' or day=='0' : # 만약 입력을 종료하고 싶으면 명령어 입력하여 입력 중단
      print(">_ Force Stop")
      break

    if day=='show': # 입력 중간에 파일을 보고 싶으면 해당 명령어 입력하여 입력 내용 확인
      display(tmp)
      break
    if day=='delete':
      print("Valid date: ", tmp["date"])
      d=int(input("Input date to delete: "))
      tmp=tmp[tmp["date"] != d]
      tmp.to_csv("/content/drive/MyDrive/Colab_Notebooks/daylist.csv", index=False)
      continue

    if len(day)!=4: # 날짜 값을 잘못 입력했을 경우 continue 구문을 사용하여 다시 시도하도록 함.
      print(">_ Invalid Input. Please try again.\n")
      continue

    cont=input("Write What you learn: ") # 배운 내용 입력

    new_data=pd.DataFrame({
        'date': [int(day)],
        'content': [cont]
        })
    # DataFrame 으로 새로 정의

    new_data.to_csv('/content/drive/MyDrive/Colab_Notebooks/daylist.csv', mode='a', header=False)
    print()
    # mode='a' 로 하여 내용을 추가, header=False 로 새로운 컬럼 이름을 다시 쓰지 않도록 설정
```
